## 모델 불러오기

```python
# 객체 검출해보기
import torch
model = torch.hub.load('ultralytics/yolov5', 'custom', path='best.pt', force_reload=True)
```

## 이미지 불러오기

```python
# 이미지 띄우기
import cv2

img = cv2.imread('images/cat (1).png')

cv2.imshow('cat vs dog', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 이미지 예측하기

```python
result = model(img)
result.xyxyn[0] # 찾은 객체 관련 값
# x1, y1, x2, y2, confidence, class
```

## tensor 배열 → numpy 배열

```python
import numpy
detect = result.xyxyn[0].numpy() # tensor 배열 → numpy 배열
detect
```

## 픽셀값으로 변경

```python
x1,y1,x2,y2 = detect[0,0], detect[0,1], detect[0,2], detect[0,3]
img.shape
y_shape = img.shape[0]
x_shape = img.shape[1]
x1,y1,x2,y2 = map(int, (x1 * x_shape, y1 * y_shape, x2 * x_shape, y2 * y_shape)) # 픽셀값으로 변환
x1,y1,x2,y2
```

## 객체 위치에 표현하기

```python
cv2.rectangle(img, (x1,y1), (x2,y2), (0,0,255), 2)
cv2.imshow('cat vs dog', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 2개 이상의 객체 탐지

```python
## 이미지 띄우기
import cv2
import numpy
img = cv2.imread('images/cat (5).png')

result = model(img)
print(result.xyxyn[0]) # 찾은 객체 관련 값

detect = result.xyxyn[0].numpy() # tensor 배열 → numpy 배열
for i in range(len(detect)):
    x1,y1,x2,y2 = detect[i,0], detect[i,1], detect[i,2], detect[i,3]
    img.shape
    y_shape = img.shape[0]
    x_shape = img.shape[1]
    x1,y1,x2,y2 = map(int, (x1 * x_shape, y1 * y_shape, x2 * x_shape, y2 * y_shape)) # 픽셀값으로 변환
    x1,y1,x2,y2
    cv2.rectangle(img, (x1,y1), (x2,y2), (0,0,255), 2)
cv2.imshow('cat vs dog', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 동영상에 적용

```python
## 이미지 띄우기
import cv2
import numpy
cap = cv2.VideoCapture('images/dog_movie.mp4')

while cap.isOpened() :
    ret,img = cap.read()

    result = model(img)
    print(result.xyxyn[0]) # 찾은 객체 관련 값
    
    detect = result.xyxyn[0].numpy() # tensor 배열 → numpy 배열
    for i in range(len(detect)):
        x1,y1,x2,y2 = detect[i,0], detect[i,1], detect[i,2], detect[i,3]
        img.shape
        y_shape = img.shape[0]
        x_shape = img.shape[1]
        x1,y1,x2,y2 = map(int, (x1 * x_shape, y1 * y_shape, x2 * x_shape, y2 * y_shape)) # 픽셀값으로 변환
        cv2.rectangle(img, (x1,y1), (x2,y2), (0,0,255), 2)
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(img, str(detect[i,4]), (50, 50), font, 1, (255, 0, 0), 2, cv2.LINE_AA)

    if not ret:
        print('종료')
        break
            
    # 출력
    cv2.imshow('video', img)
        
    # 초당 30장 출력하고 키 입력 받음
    # 33/1000 초 딜레이
    key = cv2.waitKey(33)
        
    # 만약 1키를 눌렀다면 종료
    if key == 49 :
        print('종료')
        break
    
cap.release()
cv2.destroyAllWindows()
```
